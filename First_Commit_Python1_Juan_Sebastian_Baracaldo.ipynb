```python
#▌│█║▌║▌║RANDOMIZACION DE NUMEROS ENTEROS ▌│█║▌║▌║
import random  

lista = [] 
for _ in range (10): 
  rand_num = random.randint(5,50)  
  lista.append(rand_num)  

print(lista)

#Numeros aleatorios pares 

lista = [] 
for _ in range (20): 
  rand_num = random.randrange(0, 100, 2)  
  lista.append(rand_num)  

print(lista) 


#Numeros aleatorios multiplos de 5 

lista = [] 
for _ in range (10): 
  rand_num = random.randint(1, 10)*5  
  lista.append(rand_num)  

print(lista) 


#NUmeros secuenciales random
import random 
numeros = [i for i in range (1,21)] 
print(numeros) 
random.shuffle(numeros)
print(numeros)

```

```python
print("▌│█║▌║▌║NUMERO MAYOR DE LA LISTA V2▌│█║▌║▌║")
mi_lista = [] 
num_elementos = int(input("Cuantos elementos va a trabajar:  ")) 
for i in range(num_elementos): 
  elementos = int(input("Ingrese un numero a la lista ")) 
  mi_lista.append(elementos) 

mayor = mi_lista[0] 
for k in mi_lista:
  if (k > mayor): 
    mayor = k 

print("El numero mayor es: ",mayor)
```

```python
print("▌│█║▌║▌║NUMEROS PARES E IMPARES V1 ▌│█║▌║▌║")  
num = 0 
numero = int(input("Ingrese un numero: ")) 
if numero%2 == 0: 
  print(numero, "Este numero es par") 
else: 
  print(numero, "Este numero es impar")
```

```python
print("▌│█║▌║▌║NUMEROS PARES E IMPARES V2 ▌│█║▌║▌║")   
num = 0 
while True: 
  try:
    numero = int(input("Ingrese un numero: ")) 
  except ValueError: 
    print("Error el dato ingresado no es un numero xd") 
    continue 
  else: 
     if numero%2 == 0: 
      print(numero, "Este numero es par") 
     else: 
      print(numero, "Este numero es impar")  
      break
```

```python
#▌│█║▌║▌║SUMA DE NUMEROS PARES E IMPARES▌│█║▌║▌║ 
total = 0 
num = int(input("Porfavor ingrese un numero")) 
numeros_pares = [] 

for i in range(0, num + 1): 
  if i %2 == 1: #para trabajar con los impares solo cambiariamos el 0 a un 1 en el if
    numeros_pares.append(i) 
    total +=i 

print(f"Todos los numeros pares desde 0 hasta {num} es: {numeros_pares}") 
print("****************************") 
print("la suma de los numeros pares desde 0 hasta {} is {}".format(num, total))
```

```python
 #▌│█║▌║▌║CONTADOR DE DIGITOS  ▌│█║▌║▌║
dato_usuario = int(input("Ingrese su numero: "))  
dato_usuario = abs(dato_usuario) 

print("Tu numero original es: {}".format(dato_usuario)) 

cont = 0 
while dato_usuario > 0:
  dato_usuario //= 10 
  cont += 1 

print("La cantidad de digitos en su numero son: {}".format(cont))

```
