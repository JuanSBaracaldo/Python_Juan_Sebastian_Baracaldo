```python
#TEMPERATURA MEDIA (falta corregir)
all_days =int(input("De cuantos dias desea tomar la temperatura: ")) 
total = 0 
temp = [] 


for i in range(all_days): 
  next_day = int(input(f"Temperatura del dia {(i +1)}: ")) 
  temp.append(next_day) 
  total += temp[i] 

average_temp = round(total/all_days, 2) 
print("La temperatura mas alta es: {}".format(average_temp)) 

above_counter = 0 
list_nums_above = [] 

for x in temp: 
  if x > average_temp: 
    list_nums_above.append(x) 
    above_counter += 1 

print (f"{above_counter} dias sobre la media: {list_nums_above} ")
```
